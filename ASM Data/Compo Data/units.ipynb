```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('high_pro.csv')

# Display the first few rows of the dataframe
df.head()
```

```python
# Rename the column
df.rename(columns={'Hardness, Rockwell B': 'HV (GPa)'}, inplace=True)

# Apply the conversion formula
df['HV (GPa)'] = df['HV (GPa)'].apply(lambda HRB: (23.0834 * HRB - 280.98)/100 if not pd.isnull(HRB) else HRB)

# Display the updated dataframe
df.head()
```

```python
# Rename the column
df.rename(columns={'Hardness, Rockwell C': 'HV1 (GPa)'}, inplace=True)

# Apply the conversion formula
df['HV1 (GPa)'] = df['HV1 (GPa)'].apply(lambda HRC: (60.14 * HRC + 219.714)/100 if not pd.isnull(HRC) else HRC)

# Display the updated dataframe
df.head()
```

```python
df.to_csv('data.csv', index=False)
```

```python
df2 = pd.read_csv('data.csv')
```

```python
# Rename the columns
df2.rename(columns={
    'Tensile Strength, Yield': 'Yield Strength (MPa)',
    'Tensile Strength, Ultimate': 'Ultimate Strength (MPa)'
}, inplace=True)

# Display the updated dataframe
df2.head()
```

```python
import matplotlib.pyplot as plt
# Ensure there are no NaN values in the columns used for plotting
plt.figure(figsize=(10, 6))
plt.scatter(df2['HV (GPa)'], df2['Yield Strength (MPa)'], alpha=0.7, color='blue')
plt.title('HV (GPa) vs Yield Strength (MPa)')
plt.xlabel('HV (GPa)')
plt.ylabel('Tensile Strength, Yield')
plt.grid(True)
plt.show()
```
