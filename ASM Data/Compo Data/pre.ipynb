```python
import pandas as pd
df = pd.read_csv("high.csv")
df.head()
```

```python
# Extend the keywords list to include element names
keywords = ["Grade","Tensile", "Rockwell", "Element", "Carbon", "Chromium", "Iron", "Manganese", "Nickel", "Nitrogen", "Phosphorus", "Silicon", "Sulfur", "Aluminum", "Copper", "Molybdenum", "Vanadium", "Cobalt", "Tantalum", "Titanium", "Potassium", "Tungsten", "Zirconium", "Selenium", "Boron", "Niobium"]

# Filter columns based on the required keywords
filtered_columns = [col for col in df.columns if any(keyword in col for keyword in keywords)]

# Keep only the filtered columns
df_filtered = df[filtered_columns]

# Display the filtered DataFrame
df_filtered.head()
```

```python

df_filtered.columns = df_filtered.columns.str.strip().str.rstrip(' ')
df_filtered.to_csv("high_pro.csv", index=False)
```

```python
# Merge columns with the same name by taking the first non-null value
df_filtered = df_filtered.groupby(df_filtered.columns, axis=1).first()

# Display the updated DataFrame
df_filtered.head()
```
